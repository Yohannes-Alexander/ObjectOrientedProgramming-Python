```python
import copy

class Vehicle():
    
    def start(self):
        print('Vehicle start code')
    
    def stop(self):
        print('Vehicle stop code')

class Car(Vehicle):
    
    def setNbOfSeats(self, nbOfSeats):
        self.nbOfSeats = nbOfSeats
    
    def getNbOfSeats(self):
        return self.nbOfSeats
    
print('1')
car1 = Car()
car1.start()
car1.setNbOfSeats(4)
print("The number of seats is",car1.getNbOfSeats())
car1.stop()

print('2')
vehicle1 = copy.copy(car1)
vehicle1.__class__ = Vehicle
vehicle1.__dict__ = car1.__dict__

vehicle1.start()

vehicle1.stop()

print('3')
car2 = copy.copy(vehicle1)
car2.__class__ = Car
car2.__dict__ = vehicle1.__dict__

car2.start()
car2.setNbOfSeats(4)
print("The number of seats is",car2.getNbOfSeats())
car2.stop()


```

```python
import copy

class Vehicle():
    
    def start(self):
        print('Vehicle start code')
    
    def stop(self):
        print('Vehicle stop code')

class Car(Vehicle):
    
    def setNbOfSeats(self, nbOfSeats):
        self.nbOfSeats = nbOfSeats
    
    def getNbOfSeats(self):
        return self.nbOfSeats
     
    def start(self):
        print('Car start Code')
    

print('1')
car1 = Car()
car1.start()
car1.setNbOfSeats(4)
print("The number of seats is",car1.getNbOfSeats())
car1.stop()

print('2')
vehicle1 = copy.copy(car1)
vehicle1.__class__ = Vehicle
vehicle1.__dict__ = car1.__dict__

vehicle1.start()

vehicle1.stop()

print('3')
car2 = copy.copy(vehicle1)
car2.__class__ = Car
car2.__dict__ = vehicle1.__dict__

car2.start()
car2.setNbOfSeats(4)
print("The number of seats is",car2.getNbOfSeats())
car2.stop()
```

```python
import copy

class BendaBulat():
    
    def __init__(self, radius, warna):
        self.radius = radius
        self.warna = warna
    
    def getRadius(self):
        return self.radius
    
    def getWarna(self):
        return self.warna
    
    def display(self):
        print('The radius is',self.getRadius())
        print('The color is',self.getWarna())

class Tabung(BendaBulat):
    
    def __init__(self, radius, warna, tinggi):
        super().__init__(radius, warna)
        self.tinggi = tinggi
    
    def getTinggi(self):
        return self.tinggi
    
    def display(self):
        print('The radius is',self.getRadius())
        print('The color is',self.getWarna())
        print('The height is',self.getTinggi())
        print('The surface area is',2*3.14*self.radius*(self.radius+self.tinggi))

print('1')
tb1 = Tabung(3,4,5)
tb1.display()
print('2')
bb1 = copy.copy(tb1)
bb1.__class__ = BendaBulat
bb1.__dict__ = tb1.__dict__
bb1.display()
print('3')
tb2 = copy.copy(bb1)
tb2.__class__ = Tabung
tb2.__dict__ = bb1.__dict__
tb2.display()
```

```python
import copy

class BendaPersegi():
    
    def __init__(self, panjang = None, lebar = None, warna = None):
        if(panjang!=None and lebar!=None and warna != None):
            self.panjang = panjang
            self.lebar = lebar
            self.warna = warna
        elif(panjang!=None and lebar!=None):
            self.panjang = panjang
            self.lebar = lebar
            self.warna = 'Red'
    
    def getPanjang(self):
        return self.panjang
    
    def getLebar(self):
        return self.lebar
    
    def getWarna(self):
        return self.warna
    
    def display(self):
        print('The length is',self.getPanjang())
        print('The wide is ',self.getLebar())
        print('The color is',self.getWarna())

class Balok(BendaPersegi):
    
    def __init__(self, panjang=None, lebar=None, warna=None, tinggi=None):
        if(panjang!=None and lebar!=None and warna != None and tinggi !=None):
            super().__init__(panjang, lebar, warna)
            self.tinggi = tinggi
        elif(panjang!=None and lebar!=None and tinggi!=None):
            super().__init__(panjang, lebar)
            self.tinggi = tinggi
    
    def getTinggi(self):
        return self.tinggi
    
    def display(self):
        print('The length is',self.getPanjang())
        print('The wide is ',self.getLebar())
        print('The color is',self.getWarna())
        print('The height is',self.getTinggi())
    
    def volume(self):
        print('The Volume is',self.panjang*self.lebar*self.tinggi)

        
print('1')
tb1 = Balok(panjang = 3,lebar = 4,tinggi = 5)
tb1.display()
tb1.volume()
print('2')
bb1 = copy.copy(tb1)
bb1.__class__ = BendaPersegi
bb1.__dict__ = tb1.__dict__
bb1.display()
print('3')
tb2 = copy.copy(bb1)
tb2.__class__ = Balok
tb2.__dict__ = bb1.__dict__
tb2.display()
tb2.volume()
```
