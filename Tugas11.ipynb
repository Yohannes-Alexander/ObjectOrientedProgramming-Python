```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, holidayEntitlement, employeeName, employeeDept, lengthOfService):
        self.holidayEntitlement = holidayEntitlement
        self.employeeName = employeeName
        self.employeeDept = employeeDept
        self.lengthOfService = lengthOfService
    
    def getName(self):
        return self.employeeName
    
    def getDepartment(self):
        return self.employeeDept
    
    def getLengthOfService(self):
        return self.lengthOfService
    
    def getHolidays(self):
        return self.holidayEntitlement
    
    def toString(self):
        return("Nama:"+self.getName()+"\nDepartement:"+self.getDepartment()+"\nHoliday:"+str(self.getHolidays())+"\nLengthOfService:"+str(self.getLengthOfService()))
    
    @abstractmethod
    def displayDetails(self):
        pass

class Technician(Employee):
    def __init__(self, holidayEntitlement, employeeName, employeeDept, lengthOfService, holidays):
        super().__init__(holidayEntitlement, employeeName, employeeDept, lengthOfService)
        self.holidays = holidays
    
    def getHolidays(self):
        return self.holidays
    
    def toString(self):
        return("Nama:"+self.getName()+"\nDepartement:"+self.getDepartment()+"\nHoliday Entitlement:"+str(super().getHolidays())+"\nHoliday:"+str(self.getHolidays())+"\nLengthOfService:"+str(self.getLengthOfService()))
    
    def displayDetails(self):
        print("Technician\n"+self.toString())
    
class Manager(Employee):
    def __init__(self, holidayEntitlement, employeeName, employeeDept, lengthOfService, holidays):
        super().__init__(holidayEntitlement, employeeName, employeeDept, lengthOfService)
        self.holidays = holidays
        
    def getHolidays(self):
        return self.holidays
            
    def toString(self):
        return("Nama:"+self.getName()+"\nDepartement:"+self.getDepartment()+"\nHoliday Entitlement:"+str(super().getHolidays())+"\nHoliday:"+str(self.getHolidays())+"\nLengthOfService:"+str(self.getLengthOfService()))
    
    def displayDetails(self):
        print("Manager\n"+self.toString())
        
manager1 = Manager(3,'Yohannes','IT',5,1)
manager1.displayDetails()
print()

technician1 = Technician(3,'Yohannes','Operation',10,2)
technician1.displayDetails()
```

```python
from abc import ABC, abstractmethod
import math

class RoundShape(ABC):
    def __init__(self, radiusOfCircle, xCenter, yCenter):
        self.radiusOfCircle = radiusOfCircle
        self.objC = self.Center(xCenter,yCenter)
    
    class Center():
        def __init__(self,xCenter,yCenter):
            self.xCenter = xCenter 
            self.yCenter = yCenter
            
    @abstractmethod
    def area(self):
        pass
    

class Sphere(RoundShape):
    def __init__(self, radiusOfCircle, xCenter, yCenter):
        super().__init__(radiusOfCircle, xCenter, yCenter)
    
    def area(self):
        return 4*math.pi*((self.radiusOfCircle)**2)
    
class Circle(RoundShape):
    def __init__(self, radiusOfCircle, xCenter, yCenter):
        super().__init__(radiusOfCircle, xCenter, yCenter)
    
    def area(self):
        return math.pi*((self.radiusOfCircle)**2)
    
class Tabung(RoundShape):
    def __init__(self, radiusOfCircle, xCenter, yCenter,tinggi):
        super().__init__(radiusOfCircle, xCenter, yCenter)
        self.tinggi = tinggi
    
    def area(self):
        return 2*math.pi*self.radiusOfCircle*(self.radiusOfCircle+self.tinggi)
    

sphere1 = Sphere(4,3,2)
print("The area of spehre1 is",sphere1.area())
circle1 = Circle(4,3,2)
print("The area of circle1 is",circle1.area())
tabung1 = Tabung(4,3,2,10)
print("The area of tabung1 is",tabung1.area())
```
