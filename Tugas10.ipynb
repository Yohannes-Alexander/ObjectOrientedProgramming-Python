```python
import abc

class A(abc.ABC):
    NILAI_A = 10
    
    @abc.abstractmethod
    def hitungKali(self, a, b):
        pass
    

class B:
    NILAI_B = 20
    
    def __init__(self, b):
        self.b = b
        
    def getB(self):
        return self.b
    
    def hitungKurang(self, a,b):
        return (a-b)
    
class C(A):
    def __init__(self, c):
        self.c = c
        
    def tampilBC(self):
        objB = B(2)
        print('Nilai b adalah',objB.getB())
        print('Nilai c adalah',self.c)
    
    def tampilKonstanta(self):
        objB = B(2)
        print('NILAI_A adalah',self.NILAI_A)
        print('NILAI_B adalah',objB.NILAI_B)
        
    def hitungKali(self, a, b):
        return a*b
    
objC = C(5)
objC.tampilBC()
objC.tampilKonstanta()
print("Hasil perkalian adalah ",objC.hitungKali(1,2))
        
```

```python
import abc

class A(abc.ABC):
    NILAI_A = 10
    
    @abc.abstractmethod
    def hitungKali(self, a, b):
        pass
    

class B(abc.ABC):
    NILAI_B = 20
    
    @abc.abstractmethod
    def hitungKurang(self, a,b):
        pass
    
class C(A,B,abc.ABC):
    NILAI_C = 30
    
    @abc.abstractmethod
    def hitungJumlah(self, a, b):
        pass
    
class D(C):
    def __init__(self, d=None):
        if(d!=None):
            self.d = d
        else:
            self.d = 6
        
    def tampilD(self):
        print('Nilai d adalah',self.d)
    
    def tampilKonstanta(self):
        print('NILAI_A adalah',self.NILAI_A)
        print('NILAI_B adalah',self.NILAI_B)
        print('NILAI_C adalah',self.NILAI_C)
    
    def hitungJumlah(self,a,b):
        return a+b
    
    def hitungKurang(self,a,b):
        return a-b
        
    def hitungKali(self, a, b):
        return a*b
    
objD = D(5)
objD.tampilD()
objD.tampilKonstanta()
print("Hasil perkalian adalah",objD.hitungKali(1,2))
print("Hasil perkalian adalah",objD.hitungJumlah(1,2))
print("Hasil perkalian adalah",objD.hitungKurang(1,2))
        
```
