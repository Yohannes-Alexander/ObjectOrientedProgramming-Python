```python
class overloadingSum:
    def sum(a=None,b=None,c=None,d=None):
        if(a!=None and b!=None and c!=None and d!=None):
            result = a+b+c+d
            print("a+b+c+d = ",result)
        
        elif (a!=None and b!=None and c!=None):
            result = a+b+c
            print("a+b+c = ",result)
        
        elif(a!=None and b!=None):
            result = a+b
            print("a+b = ",result)
                
        else:
            print("Tidak cukup untuk bilangan yang ingin dijumlahkan")
            
Sum = overloadingSum
Sum.sum(1,2)
Sum.sum(1,2,3)
Sum.sum(1,2,3,4)
Sum.sum()
```

```python
class overloadingBungaMajemuk:
    Mo = 100000
    r = 0.3
    n = 12
    
    def bungaMajemuk(self,Mo=None,r=None,n=None):
        if(Mo!=None and r!=None and n!=None):
            self.Mo = Mo
            self.r = r
            self.n = n
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
        
        elif(Mo!=None and r!=None):
            self.Mo = Mo
            self.r = r
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data n (jangka waktu) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
        
        elif(Mo!=None and n!=None):
            self.Mo = Mo
            self.n = n
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data r(bunga) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
            
        elif(Mo!=None):
            self.Mo = Mo
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data r(bunga) dan n(jangka waktu) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
        
        elif(r!=None):
            self.r = r
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data  Mo(Modal Awal) dan n(jangka waktu) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
        
        elif(n!=None):
            self.n = n
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data Mo(Modal Awal) dan r(bunga) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
        
        elif(r!=None and n!=None):
            self.r = r
            self.n = n
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena data Mo(Modal Awal) tidak diinput user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
            
        else:
            bungaMajemuk = self.Mo*((1+self.r)**self.n)
            print("Karena ketiga data tidak diinput oleh user, maka diambil input secara default")
            print("Bunga majemuknya adalah ",bungaMajemuk,"\n")
            


bunga = overloadingBungaMajemuk()
ModalAwal = 100000
bungaBank = 0.5
jangkaWaktu = 10
bunga = overloadingBungaMajemuk()
bunga.bungaMajemuk(ModalAwal,bungaBank,jangkaWaktu)
bunga.bungaMajemuk(None,bungaBank,jangkaWaktu)
bunga.bungaMajemuk(ModalAwal,None,None)
bunga.bungaMajemuk()
```

```python
import math

class overloadingVolumeSilinder:
    r = 5
    t = 5
    
    def volumeSilinder(self,r=None,t=None):
        if(r!=None and t!=None):
            self.r = r
            self.t = t
            volume = math.pi*self.t*(self.r**2)
            print("Volume silindernya adalah ",volume,"\n")
        elif(r!=None):
            self.r = r
            volume = math.pi*self.t*(self.r**2)
            print("Karena data t(tinggi silinder) tidak diinput oleh user, maka diinput secara default")
            print("Volume silindernya adalah ",volume,"\n")
        elif(t!=None):
            self.t = t
            volume = math.pi*self.t*(self.r**2)
            print("Karena data r(jari-jari silinder) tidak diinput oleh user, maka diinput secara default")
            print("Volume silindernya adalah ",volume,"\n")
        else:
            volume = math.pi*self.t*(self.r**2)
            print("Karena data r(jari-jari silinder) dan t(tinggi silinder) tidak diinput oleh user, maka diinput secara default")
            print("Volume silindernya adalah ",volume,"\n")
            

silinder = overloadingVolumeSilinder()
silinder.volumeSilinder(10,10)
silinder.volumeSilinder(10,None)
silinder.volumeSilinder(None,10)
silinder.volumeSilinder()
```
