```python
class B:
    def __init__(self, a, b):
        self.__a = a
        self.__b = b
    
    def getA(self):
        return self.__a
    
    def getB(self):
        return self.__b
    
    def setA(self, a):
        self.__a = a
    
    def setB(self, b):
        self.__b = b
    
    def showA(self):
        print("The value of a is",self.getA())
    
    def showB(self):
        print("The value of b is",self.getB())
    
class D(B):
    def __init__(self,a,b):
        super().__init__(a,b)
        self.__c = 0
        self.__d = 0
    
    def mul(self):
        self.__c = self.getB()*self.getA()
    
    def dev(self):
        self.__d = self.getB() / self.getA()
    
    def display(self):
        self.showA()
        self.showB()
        print("The value of c is",self.__c)
        print("The value of d is",self.__d)

        
varA = B(1.5,2.5)
print("Variabel A in class B")
varA.showA()
varA.showB()
print()

varB = D(2.0,7.5)
print("Variabel B in class D")
varB.dev()
varB.mul()
varB.display()
        
```

```python
class Student:
    def __init__(self, rollNumber=None):
        if(rollNumber!=None):
            self.__rollNumber = rollNumber
        else:
            self.__rollNumber = '000000000000'
    
    def putNumber(self):
        print("The rollNumber is ",self.__rollNumber)
        
    def setNumber(self, rollNumber):
        self.__rollNumber = rollNumber
    

class Test(Student):
    def __init__(self, rollNumber, x, y):
        super().__init__(rollNumber)
        self.__sub1 = x
        self.__sub2 = y
    
    def putMarks(self):
        print("The sub1 value is ", self.getSub1())
        print("The sub2 value is ", self.getSub2())
    
    def getSub1(self):
        return self.__sub1
    
    def getSub2(self):
        return self.__sub2
    
    def setSub1(self, x):
        self.__sub1 = x
    
    def setSub2(self, y):
        self.__sub2 = y

class Result(Test):
    def __init__(self, rollNumber, x, y):
        super().__init__(rollNumber, x, y)
        self.__total = 0
    
    def display(self):
        self.__total = self.getSub1() + self.getSub2()
        self.putMarks()
        print("The total of sub1 and sub2 is ",self.__total)
    
student1 = Result(82011633062,80,90)
print("The data of student 1")
student1.putNumber()
student1.display()
```

```python
class Shape:
    def __init__(self, aName):
        self.__aName = aName
    
    def getName(self):
        return self.__aName
    
    def calculateArea(self):
        return 0.0

import math
class Circle(Shape):
    def __init__(self, cName, r):
        super().__init__(cName)
        self.__r = r
    
    def calculateArea(self):
        return ((self.__r**2)*math.pi)
    
    def getRadius(self):
        return self.__r
    
    def setRadius(self,r):
        self.__r = r
        print("The radius of",self.getName(),"has changed into",self.__r)
    
    def display(self):
        print("The radius of",self.getName(),"is",self.getRadius())
        print("The area of",self.getName(),"is",self.calculateArea())
    
class Square(Shape):
    def __init__(self, cName, s):
        super().__init__(cName)
        self.__s = s
        
    def calculateArea(self):
        return self.__s*self.__s
    
    def getSide(self):
        return self.__s
    
    def setSide(self,s):
        self.__s = s
        print("The side of",self.getName(),"has changed into",self.__s)
    
    def display(self):
        print("The radius of",self.getName(),"is",self.getSide())
        print("The area of",self.getName(),"is",self.calculateArea())


circle = Circle("Circle1",10.0)
circle.display()
print(circle.getName())

square = Square("Square1",30.0)
square.display()
```
