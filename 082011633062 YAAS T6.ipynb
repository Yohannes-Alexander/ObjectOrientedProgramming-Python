```python
class Vektor:
    def __init__(self,element=None,size=None):
        if(element!=None and size!=None):
            self.element=element
            self.size=size
        elif(element!=None):
            self.element=element
            self.size=len(self.element)
        elif(size!=None):
            self.element=[]
            for i in range(size):
                self.element.append(0)
        else:
            print("Please ensure your parameter input and complete one of them")

    def jumlah2Vektor(self, vektor):
        vektorResult=Vektor(None,vektor.size)
        for i in range(self.size):
            vektorResult.element[i]=vektor.element[i]+self.element[i]
        
        vektorResult.size=len(vektorResult.element)
        return vektorResult
    
    def selisih2Vektor(self,vektor):
        vektorResult=Vektor(None,vektor.size)
        for i in range(self.size):
            vektorResult.element[i]=self.element[i]-vektor.element[i]
        
        vektorResult.size=len(vektorResult.element)
        return vektorResult
    
    def tampilVektor(self):
        print(self.element)
        
vektor1=Vektor([1,4,3],len([1,4,3]))
vektor1.tampilVektor()
vektor2=Vektor([1,2,7],len([1,2,7]))
vektor2.tampilVektor()
vektorResultan = vektor1.jumlah2Vektor(vektor2)
print("Jumlah vektor1 dan vektor2 adalah ")
vektorResultan.tampilVektor()
vektorResultan = vektor1.selisih2Vektor(vektor2)
print("Selisih vektor1 dan vektor2 adalah ")
vektorResultan.tampilVektor()
```

```python
class Matriks:
    def __init__(self,element,row,column):
        self.element = element
        self.row = row
        self.column = column
    
    def jumlah2Matriks(self,matriks1, matriks2):
        ResultMatriks = Matriks(matriks1.element,matriks1.row,matriks1.column)
        for i in range(matriks1.row):
            for j in range(matriks1.column):
                ResultMatriks.element[i][j]=matriks1.element[i][j]+matriks2.element[i][j]
        return ResultMatriks
    
    def pengurangan2Matriks(self, matriks1, matriks2):
        ResultMatriks = Matriks(matriks1.element,matriks1.row,matriks1.column)
        for i in range(matriks1.row):
            for j in range(matriks1.column):
                ResultMatriks.element[i][j]=matriks1.element[i][j]-matriks2.element[i][j]
        return ResultMatriks
        
    def tampilArray(self):
        for row in (self.element):
            for column in row:
                print(column ,end=" ")
            print()

Matriks1 = Matriks([[1,2,3], [1,4,5], [5,3,6]],3,3)
print("Matriks 1")
Matriks1.tampilArray()
print()
Matriks2 = Matriks([[5,1,7], [4,0,5], [1,3,9]],3,3)
print("Matriks 2")
Matriks2.tampilArray()
print()

ResultMatriks = Matriks([[0,0,0], [0,0,0], [0,0,0]],3,3)
ResultMatriks = ResultMatriks.jumlah2Matriks(Matriks1,Matriks2)
print("Penjumlahan dua matriks")
ResultMatriks.tampilArray()
print()

Matriks1 = Matriks([[1,2,3], [1,4,5], [5,3,6]],3,3)
print("Matriks 1")
Matriks1.tampilArray()
print()
Matriks2 = Matriks([[5,1,7], [4,0,5], [1,3,9]],3,3)
print("Matriks 2")
Matriks2.tampilArray()
print()

PenguranganMatriks = Matriks([[0,0,0], [0,0,0], [0,0,0]],3,3)
PenguranganMatriks = PenguranganMatriks.pengurangan2Matriks(Matriks1,Matriks2)
print("Pengurangan dua matriks")
PenguranganMatriks.tampilArray()
```
